```python
import numpy as np

initial_price = 120_000
growth_rate = 0.05
years = 5

final_price = initial_price * (1 + growth_rate) ** years
print(f"Cena mieszkania za 5 lat: {final_price:.2f} zł")
```

```python
import numpy_financial as npf

rate = 0.12 / 12
nper = years * 12

# Obliczenie wymaganej miesięcznej wpłaty (PMT):
monthly_payment = -npf.pmt(rate, nper, 0, final_price)
print(f"Musisz wpłacać miesięcznie: {monthly_payment:.2f} zł")
```

```python
import matplotlib.pyplot as plt

# Ceny mieszkań w czasie (liniowy wzrost co miesiąc)
months = np.arange(nper)
price_trend = initial_price * (1 + growth_rate) ** (months / 12)

# Skumulowana wartość lokaty w czasie:
investment_value = npf.fv(rate, months, -monthly_payment, 0)

plt.plot(months, price_trend, label='Cena mieszkania')
plt.plot(months, investment_value, label='Wartość lokaty')
plt.xlabel("Miesiące")
plt.ylabel("Wartość (zł)")
plt.legend()
plt.title("Cena mieszkania vs. oszczędności")
plt.grid(True)
plt.tight_layout()
plt.show()
```
